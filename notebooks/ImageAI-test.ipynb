```python
from imageai.Prediction import ImagePrediction
from imageai.Detection import ObjectDetection
```

```python
import os
import cv2
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
# execution_path = os.getcwd()
execution_path = '../models'

prediction = ImagePrediction()
prediction.setModelTypeAsResNet()
prediction.setModelPath(os.path.join(execution_path, "resnet50_weights_tf_dim_ordering_tf_kernels.h5"))
# prediction.loadModel()
# prediction.loadModel()
# prediction.loadModel(prediction_speed="fast")
# prediction.loadModel(prediction_speed="faster")
prediction.loadModel(prediction_speed="fastest")
```

```python
start = datetime.now()

predictions, probabilities = prediction.predictImage(os.path.join('./', "1.jpg"), result_count=5 )
    
end = datetime.now()

print(end - start)
    
image = cv2.imread("1.jpg")
image = np.stack([image[:, :, 2], image[:, :, 1], image[:, :, 0]], 2)

plt.figure()
plt.imshow(image)
plt.show()

for eachPrediction, eachProbability in zip(predictions, probabilities):
    print(eachPrediction , " : " , eachProbability)
```

```python
execution_path = "../models"

detector = ObjectDetection()
detector.setModelTypeAsYOLOv3()
detector.setModelPath(os.path.join(execution_path , "yolo.h5"))
detector.loadModel()

tiny_detector = ObjectDetection()
tiny_detector.setModelTypeAsTinyYOLOv3()
tiny_detector.setModelPath(os.path.join(execution_path , "yolo-tiny.h5"))
tiny_detector.loadModel()
```

```python
start = datetime.now()

detections = detector.detectObjectsFromImage(input_image=os.path.join('./' , "image2.jpg"), output_image_path=os.path.join('./' , "image2new.jpg"), minimum_percentage_probability=30)

end = datetime.now()

print(end - start)

image = cv2.imread("image2.jpg")
image = np.stack([image[:, :, 2], image[:, :, 1], image[:, :, 0]], 2)

image2 = cv2.imread("image2new.jpg")
image2 = np.stack([image2[:, :, 2], image2[:, :, 1], image2[:, :, 0]], 2)

fig, axes = plt.subplots(1, 2, figsize=(14, 7))
axes[0].imshow(image)
axes[1].imshow(image2)
plt.show()

for eachObject in detections:
    print(eachObject["name"] , " : ", eachObject["percentage_probability"], " : ", eachObject["box_points"] )
    print("--------------------------------")
```
