```python
image = face_recognition.load_image_file("itzy.jpg")
face_locations = face_recognition.face_locations(image)
print(face_locations)

fig, ax = plt.subplots(figsize=(18, 8))
ax.imshow(image)

for x1, y1, x2, y2 in face_locations:
    rect = patches.Rectangle((y1, x1), y2 - y1, x2 - x1, edgecolor='r', facecolor="none", linewidth=2)
    ax.add_patch(rect)

plt.show()
```

```python
# https://www.learnopencv.com/read-write-and-display-a-video-using-opencv-cpp-python/

# Create a VideoCapture object and read from input file
# If the input is the camera, pass 0 instead of the video file name
cap = cv2.VideoCapture(mp4_files[0])

# Check if camera opened successfully
if (cap.isOpened()== False): 
    print("Error opening video stream or file")

frame_count = 0

for i in range(6000):
    ret, frame = cap.read()
    
# Read until video is completed
while(cap.isOpened()):
    # Capture frame-by-frame
    
    
    for i in range(6000):
        ret, frame = cap.read()
    
    if ret == True:
#         cv2.imshow('Frame',frame)
#         time.sleep(0.01)
        faces = face_locations = face_recognition.face_locations(frame)
    
        if len(faces) > 0:
            fig, ax = plt.subplots(figsize=(6, 4))
            
            frame = np.stack((frame[:, :, 2], frame[:, :, 1], frame[:, :, 0]), axis=2)
            
            ax.imshow(frame)

            for x1, y1, x2, y2 in face_locations:
                rect = patches.Rectangle((y1, x1), y2 - y1, x2 - x1, edgecolor='r', facecolor="none", linewidth=2)
                ax.add_patch(rect)

            plt.show()
            
        print("Checked")

        # Press Q on keyboard to  exit
        if cv2.waitKey(25) & 0xFF == ord('q'):
            break

    else: 
        break

cap.release()

# Closes all the frames
cv2.destroyAllWindows()
```
